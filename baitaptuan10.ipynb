```python
#W10A10
def count(n, arr):
    count = 0
    for i in arr:
        if i == n:
            count += 1
    if count == 1:
        return True
    else:
        return False


n = int(input())
nums = list(map(int, input().split()))
if n != len(nums):
    print('vui long nhap dung so luong')
else:
    for i in nums:
        if count(i, nums):
            sum = 0
            for k in nums:
                sum += k
            print(sum)
            break
        elif not count(i, nums):
            print('Mang khong hop le')
            break
```
