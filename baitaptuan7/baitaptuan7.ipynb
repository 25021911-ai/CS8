```python
#W7A1
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    while left <= right :
        mid = ( left + right ) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
arr = list(map(int, input().split()))
target = int(input())
print(binary_search(arr, target))

    
    
    
```

```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
arr = list(map(int, input().split()))
print(bubble_sort(arr))
print(selection_sort(arr))
```

```python
#W7A4
def count_frequency(arr):
    frequences_num = {}
    for num in arr:
        if num in frequences_num:
            frequences_num[num] += 1
        else:
            frequences_num[num] = 1
    max_num = None
    max_count = -1
    for num, count in frequences_num.items():
        if count > max_count:
            max_count = count
            max_num = num
    return max_num
arr = list(map(int, input().split()))
n = count_frequency(arr)
print(f'{n} xuat hien nhieu nhat som nhat')


    
        

```

```python
#W7A5
def sum_count(arr, x):
    n = len(arr)
    L = []
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] + arr[j] == x :
                L.append((arr[i],arr[j]))
    return L
arr = list(map(int, input().split()))
c = int(input())
result = sum_count(arr, c)
print(len(result))

```

```python
#W7A7
L = list(map(int, input().split()))
def manual_bisect_left(tails, x):
    low = 0
    high = len(tails)
    while low < high:
        mid = (low + high) // 2
        if tails[mid] < x:
            low = mid + 1
        else:
            high = mid
    return low
def find_string(arr):
    tails = []
    for x in arr:
        idx = manual_bisect_left(tails, x)
        if idx == len(tails):
            tails.append(x)
        else:
            tails[idx] = x
    return len(tails)
print(find_string(L))

            

        
```

```python
#W7A10
def find_max_mod(nums, m):
    max_mod = 0
    current_sum = 0
    seen = [0] 
    for x in nums:
        current_sum = (current_sum + x) % m
        if current_sum > max_mod:
            max_mod = current_sum
        low = 0
        high = len(seen) - 1
        idx = -1
        while low <= high:
            mid = (low + high) // 2
            if seen[mid] > current_sum:
                idx = mid
                high = mid - 1
            else:
                low = mid + 1       
        if idx != -1:
            prev = seen[idx]
            potential_sum = (current_sum - prev + m) % m
            if potential_sum > max_mod:
                max_mod = potential_sum
        insert_idx = 0
        while insert_idx < len(seen) and seen[insert_idx] < current_sum:
            insert_idx += 1
        seen.insert(insert_idx, current_sum)
    return max_mod
lst = list(map(int, input("Nhập mảng (cách nhau bằng khoảng trắng): ").split()))
n = int(input("Nhập m số nguyên m: "))
print("Kết quả lớn nhất:", find_max_mod(lst, n))
```

```python
#W7A13
def find_lhs_optimized(arr):
    counts = {}
    max_length = 0
    for x in arr:
        counts[x] = counts.get(x, 0) + 1
        if (x - 1) in counts:
            max_length = max(max_length, counts[x] + counts[x-1])
        if (x + 1) in counts:
            max_length = max(max_length, counts[x] + counts[x+1])
    return max_length
lst = list(map(int, input().split()))
print(find_lhs_optimized(lst))
```
