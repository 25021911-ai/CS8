```python
#W5A1
def max(a, b):
    if a > b:
        print(a)
    else:
        print(b)
max(1, 2)

```

```python
#W5A2
def swap(a, b):
    tuple = (b, a)
    return tuple
swap(1, 2)
```

```python
#W5A3
def prime(n):
    for i in range(2, n):
        if n%i==0:
            return False
    return True
prime(11)


        
```

```python
#W5A4
def perfect_number(n):
    sum=0
    for i in range (1, n):
        if n%i==0:
            sum+=i
    if sum==n:
        return True
    else:
        return False
perfect_number(6)
```

```python
#W5A5
def position_of_digit(l, n):
    for i in range(len(l)):
        if l[i]==n:
            return i+1
position_of_digit([1, 2, 3, 4, 5, -6, 7, 8, 19, 1, 2, 3, 4, 5], 5)
```

```python
#W5A6
def factorial(n):
    result=1
    for i in range(1, n+1):
        result=result*i
    return result
factorial(20)
```

```python
#W5A7
def math(a, b, op):
    if op=='+':
        return a+b
    elif op=='-':
        return a-b
    elif op=='*':
        return a*b
    elif op=='/':
        return a/b
math(2, 3, '+')
```

```python
#W5A8
def hammstring(n1, n2):
    a=n1^n2
    return bin(a).count('1')
hammstring(123456, 32321)
```

```python
#W5A9
def sum_of_digits(n):
    sum=0
    for digit in str(n):
        sum+=int(digit)
    return sum
sum_of_digits(1234)
```

```python
#W5A10 
def isomorphism(s1, s2):
    if len(s1)!=len(s2):
        return False
    map_s_to_t = {}
    map_t_to_s = {}
    for c1, c2 in zip(s1, s2):
        if c1 in map_s_to_t:
            if map_s_to_t[c1] != c2:
                return False
        else:
            if c2 in map_t_to_s:
                return False
            map_s_to_t[c1] = c2
            map_t_to_s[c2] = c1
    return True
isomorphism("show", "same")
```
