```python
#W8A2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)

    def days_in_month(self):
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif self.month in [4, 6, 9, 11]:
            return 30
        elif self.month == 2:
            return 29 if self.is_leap_year() else 28
        return 0

    def next_date(self):
        max_days = self.days_in_month()
        
        if self.day > max_days or self.day < 1 or self.month < 1 or self.month > 12:
            return "Invalid Date"

        if self.day < max_days:
            self.day += 1
        else:
            self.day = 1
            if self.month == 12:
                self.month = 1
                self.year += 1
            else:
                self.month += 1
        
        return f'{self.day:02}/{self.month:02}/{self.year}'

if __name__ == "__main__":
    try:
        s = input("Nhập ngày tháng năm (dd/mm/yyyy): ")
        d, m, y = map(int, s.split('/'))
        
        my_date = Date(d, m, y)
        
        result = my_date.next_date()
        
        if result == "Invalid Date":
            print("Invalid")
        else:
            print(result)
            
    except ValueError:
        print("Invalid")    
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return (self.w + self.h) * 2

    def scale(self, k):
        self.w = self.w * k
        self.h = self.h * k

if __name__ == "__main__":
    try:
        w, h, k = map(float, input().split())
        
        rect = Rectangle(w, h)
        
        rect.scale(k)
        
        print(f"{rect.area():.2f} {rect.perimeter():.2f}")

    except ValueError:
        pass
```

```python
#W8A7
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số không được bằng 0")
        
        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common
        
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def add(self, other):
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def sub(self, other):
        new_num = self.num * other.den - other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def mul(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)

    def div(self, other):
        new_num = self.num * other.den
        new_den = self.den * other.num
        return Fraction(new_num, new_den)

    def __str__(self):
        return f"{self.num}/{self.den}"

if __name__ == "__main__":
    try:
        parts = input().split()
        
        a = int(parts[0])
        b = int(parts[1])
        op = parts[2]
        c = int(parts[3])
        d = int(parts[4])

        f1 = Fraction(a, b)
        f2 = Fraction(c, d)

        result = None
        if op == '+':
            result = f1.add(f2)
        elif op == '-':
            result = f1.sub(f2)
        elif op == '*':
            result = f1.mul(f2)
        elif op == '/':
            result = f1.div(f2)
        
        print(result)

    except ValueError:
        print("Lỗi nhập liệu")
```

```python
#W8A9:
class Student:
    def __init__(self, name):
        self.name = name
        self.lst = []
    def add_score(self, subject, score):
        result = {'subject': subject, 'score': score}
        self.lst.append(result)
        print('Da them hoc sinh thanh cong')
    def avg_score(self):
        sum = 0
        for student in self.lst:
            sum += student['score']
            avg = round(sum/(len(self.lst)), 2)
        self.avg = avg
        return self.avg
    def rank(self):
        if self.avg >= 8:
            print( "Excellent" )
        elif 6.5 <= self.avg < 8:
            print('Good')
        elif 5 <= self.avg < 6.5:
            print('Average')
        else:
            print('Poor')
if __name__ == '__main__':
    name = input()
    n = int(input())
    student = Student(name)
    for i in range(n):
        data = input().split()
        subject = data[0]
        score = float(data[1])
        student.add_score(subject, score)
        student.avg_score()
    print(f"Học sinh: {student.name}")
    print(f"Điểm trung bình: {student.avg_score()}")
    print("Xếp loại: ", end="")
    student.rank()
        



        
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.bookshelf = []
    def add(self, book):
        self.bookshelf.append(book)
    def count_by_author(self, name):
        count = 0
        for book in self.bookshelf:
            if book.author == name:
                count += 1
        return count
    def count_by_year(self, y):
        same_year_books = []
        for book in self.bookshelf:
            if book.year == y:
                same_year_books.append(book)
        return same_year_books
library = Library()
n = int(input())
for i in range(n):
    line = input().strip()
    parts = line.split(' ', 1)
    command = parts[0]
    
    if command == 'ADD':
        info = parts[1].split(';')
        title = info[0]
        author = info[1]
        year = info[2]
        
        new_book = Book(title, author, year)
        library.add(new_book)
        
    elif command == 'COUNT':
        author_name = parts[1]
        result = library.count_by_author(author_name)
        print(result)
        
    elif command == 'COUNTYEAR':
        year = int(parts[1])
        result = library.count_by_year(year)
        print(result)






```

```python
#W8A11
class coordinate:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def add(self, other):
        new_x = self.x + other.x
        new_y = self.y + other.y
        new_z = self.z + other.z
        return coordinate(new_x, new_y, new_z)
    def sub(self, other):
        new_x = self.x - other.x
        new_y = self.y - other.y
        new_z = self.z - other.z
        return coordinate(new_x, new_y, new_z)
    def direct_accum(self, other):
        new_x = (self.y * other.z) - (other.y * self.z)
        new_y = (self.z * other.x) - (other.z * self.x)
        new_z = (self.x * other.y) - (other.x * self.y)
        return coordinate(new_x, new_y, new_z)
    def product(self, other):
        return ( (self.x * other.x) + (self.y * other.y) + (self.z * other.z) )
    def length(self):
        return ( ((self.x)**2) + ((self.y)**2) + ((self.z)**2) )**0.5
    def __str__(self):
        return f'({self.x}, {self.y}, {self.z})'

x1, y1, z1 = map(int, input().split())
x2, y2, z2 = map(int, input().split())
x3, y3, z3 = map(int, input().split())
x4, y4, z4 = map(int, input().split())
A = coordinate(x1 ,y1, z1)
B = coordinate(x2, y2, z2)
C = coordinate(x3, y3, z3)
D = coordinate(x4, y4, z4)
vectoAB = B.sub(A)
vectoAC = C.sub(A)
vectoBC = C.sub(B)
vectoBD = D.sub(B)
n1 = vectoAB.direct_accum(vectoAC)
n2 = vectoBC.direct_accum(vectoBD)
a = n1.product(n2)
b = n1.length() * n2.length()
cos = a/b
import math
angle_radian = math.acos(cos)
angle_degree = round( math.degrees(angle_radian), 2)
print(angle_degree)





```

```python
#W8A12
class  ComplexNumber:
    def __init__(self, realpart, imaginarypart):
        self.realpart = realpart
        self.imaginarypart = imaginarypart
    def complexnumbermultiply(self, other):
        new_realpart = ( self.realpart * other.realpart) - (self.imaginarypart * other.imaginarypart)
        new_imaginarypart = (self.realpart * other.imaginarypart) + (self.imaginarypart * other.realpart)
        return ComplexNumber(new_realpart, new_imaginarypart)
    def __str__(self):
        return f'{self.realpart} {self.imaginarypart}'
a, b = map(int, input().split())
c, d = map(int, input().split())
num1 = ComplexNumber(a, b)
num2 = ComplexNumber(c, d)
print(num1.complexnumbermultiply(num2))
```

```python
#W8A13
class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3
    def get_area(self):
        p = 0.5 * (self.edge1 + self.edge2 + self.edge3)
        area = (p*(p - self.edge1)*(p - self.edge2)*(p - self.edge3))**0.5
        return area
a, b, c = map(int, input().split())
if a < 0 or b < 0 or c < 0:
    print('invalid')
elif a + b <= c or a + c <= b or b + c <= a:
    print('invalid')
else:
    triangle = Triangle(a, b, c)
    print(round(triangle.get_area(), 2))
    
```
